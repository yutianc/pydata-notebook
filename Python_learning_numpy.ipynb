```python
import numpy as np

lst = [[1,2,3],[4,5,6]]
print(type(lst))

nplst = np.array(lst)
print(type(nplst))

nplst_float = np.array(lst, np.float)
print(type(nplst_float))

print(nplst_float.shape)
print(nplst_float.ndim)
print(nplst_float.dtype)
print(nplst_float.itemsize)
print(nplst_float.size)
```

### 常用数组
#### 用 numpy.random 生成各类常用数组，各类分布

```python
print(np.zeros([2,13]))
print('随机数组')
print(np.random.rand(2,5))
print('随机数组-整数')
print(np.random.randint(1,10,size=(2,5)))
print('随机数组—正态分布随机数')
print(np.random.randn(2,5))
print('随机数组—从给定选择中随机选择 choice')
print(np.random.choice([2,4,7,9,12],size=(2,11)))

```

#### 数组相关的操作

```python
nplist2 = np.arange(0,100).reshape([5,20])
print(nplist2)

print(np.exp(nplist2))

print(np.sin(nplist2))
```

```python

print(np.sum(nplist2, axis=1))

print(np.max(nplist2, axis=1))

print('np.dot')
print(np.dot)
```

### 线性方程组 和 矩阵运算

```python
from numpy.linalg import *

## 单位矩阵
print(np.eye(4))
```

```python
lst = np.array([[1,2],
               [3,4]])

# 装置矩阵
print(lst.transpose())

#逆矩阵
print(inv(lst))
```
